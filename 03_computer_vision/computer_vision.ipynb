````python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context

from keras.datasets import mnist

df = mnist.load_data('minst.db')
train,test = df
X_train, y_train = train
X_test, y_test = test

import matplotlib.pyplot as plt
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000)
for i in range(5): 
    print(y_test[i],'\n') # Print the label
    print(X_test[i],'\n') # Print the matrix of values
    plt.contourf(np.rot90(X_test[i].transpose())) # Make a contour plot of the matrix values
    plt.show()
````

````python
print(X_train.shape)
print(X_test.shape)
print(X_train[0].shape)
print(X_test[0].shape)
print(y_train.shape)
print(y_test.shape)
````

````python
X_train_reshaped = X_train.reshape(X_train.shape[0], 28*28)
X_test_reshaped = X_test.reshape(X_test.shape[0], 28*28)

print(X_train_reshaped.shape)
print(X_test_reshaped.shape)
````

````python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

def find_optimal_k(X_train, y_train, X_test, y_test):
    accuracies = []
    k_values = range(1, 50)
    for k in k_values: 
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train, y_train)
        predictions = knn.predict(X_test)
        accuracy = accuracy_score(y_test, predictions)
        accuracies.append(accuracy)
        print(f"k={k}, accuracy={accuracy}")
    
    optimal_k = accuracies.index(max(accuracies)) + 1

    plt.plot(k_values, accuracies, marker='o')
    plt.xlabel('k')
    plt.ylabel('Accuracy')
    plt.title(f'Optimal k: {optimal_k}')
    plt.show()

    return optimal_k

optimal_k = find_optimal_k(X_train_reshaped, y_train, X_test_reshaped, y_test)
print(f"Optimal number of neighbors: {optimal_k}")
````

````python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import pandas as pd

knn = KNeighborsClassifier(n_neighbors=optimal_k) 
predictor = knn.fit(X_train_reshaped, y_train)
y_hat = predictor.predict(X_test_reshaped)

accuracy = knn.score(X_test_reshaped, y_test)
print('Accuracy: ', accuracy)

conf_matrix = pd.crosstab(y_test, y_hat, rownames=['Actual'], colnames=['Predicted'])
print(conf_matrix)
````
